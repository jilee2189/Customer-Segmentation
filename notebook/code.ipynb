### Clustering

Now that I have reduced the attributes to three dimensions, I will be performing clustering via Agglomerative clustering. Agglomerative clustering is a hierarchical clustering method. It involves merging examples until the desired number of clusters is achieved.

Steps involved in the Clustering

- Elbow Method to determine the number of clusters to be formed
- Clustering via Agglomerative Clustering
- Examining the clusters formed via scatter plot

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Range of clusters to try
range_n_clusters = list(range(2, 11))   # usually silhouette not defined for k=1

silhouette_avg_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_ds)
    silhouette_avg = silhouette_score(scaled_ds, cluster_labels)
    silhouette_avg_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the average silhouette_score is {silhouette_avg:.3f}")

# Plotting
plt.figure(figsize=(8,5))
plt.plot(range_n_clusters, silhouette_avg_scores, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Average silhouette score")
plt.title("Silhouette Method For Optimal k")
plt.show()
```

The above cell indicates that four will be an optimal number of clusters for this data. Next, we will be fitting the Agglomerative Clustering Model to get the final clusters.

```python
#Initiating the Agglomerative Clustering model 
AC = AgglomerativeClustering(n_clusters=4)
# fit model and predict clusters
yhat_AC = AC.fit_predict(scaled_ds)
scaled_ds["Clusters"] = yhat_AC
#Adding the Clusters feature to the orignal dataframe.
data["Clusters"]= yhat_AC
```

### Recommendation System

we can recommend what to promote and where to promote it (web / catalog / store / deals) using your existing clusters. Below is a clean, copy-paste pipeline that:

- Profiles each cluster’s top product categories and preferred channels
- Builds a hybrid recommender (cluster prior + item-based CF + recency)
- Returns per-user Top-N categories with ideal marketing venues and short “why” notes
- Also outputs per-cluster recommendations for campaign planning

```python
from sklearn.metrics.pairwise import cosine_similarity
# ---------- 0) Setup ----------
d = data.copy()

# Items = product categories (columns beginning with "Mnt")
item_cols = ["Wines","Fruits","Meat","Fish","Sweets","Gold"]
assert item_cols, "No item columns found. Expect columns like 'MntWines', 'MntMeatProducts', etc."

# Channels (use what's available)
channel_cols = [c for c in ["NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases"] if c in d.columns]
assert channel_cols, "No channel columns found (expected one of NumWebPurchases/NumCatalogPurchases/NumStorePurchases/NumDealsPurchases)."

# Optional promo acceptance:
promo_cols = [c for c in d.columns if c.startswith("AcceptedCmp")]
if "Response" in d.columns: 
    promo_cols += ["Response"]

# Safety
d["Clusters"] = d["Clusters"].astype(int)
if "Recency" not in d.columns:
    raise ValueError("Need Recency (days since last engagement).")

# ---------- 1) Cluster profiles (products + channels) ----------
# Product preference per cluster (mean spend, normalized to proportions)
clust_item_mean = d.groupby("Clusters")[item_cols].mean()
clust_item_prop = clust_item_mean.div(clust_item_mean.sum(axis=1).replace(0,1), axis=0).fillna(0)

# Channel preference per cluster (mean counts, normalized to proportions)
clust_channel_mean = d.groupby("Clusters")[channel_cols].mean()
clust_channel_prop = clust_channel_mean.div(clust_channel_mean.sum(axis=1).replace(0,1), axis=0).fillna(0)

# (Optional) Promo responsiveness per cluster
if promo_cols:
    clust_promo_rate = d.assign(_promo=d[promo_cols].sum(axis=1)).groupby("Clusters")["_promo"].mean()
else:
    clust_promo_rate = pd.Series(index=sorted(d["Clusters"].unique()), dtype=float)

# Helper: top-N lists per cluster (for dashboards / campaigns)
def top_items_for_cluster(cid, topn=5):
    return clust_item_prop.loc[cid].sort_values(ascending=False).head(topn)

def top_channels_for_cluster(cid, topn=3):
    return clust_channel_prop.loc[cid].sort_values(ascending=False).head(topn)

# ---------- 2) Personalization via item-based CF + cluster prior + recency ----------
# Implicit feedback matrix (weighted by spend)
R_w = np.log1p(d[item_cols].fillna(0).astype(float)).to_numpy()   # users × items

# Item similarity (cosine)
S_items = cosine_similarity(R_w.T)
np.fill_diagonal(S_items, 0.0)

# Raw CF score for each user (users × items)
cf_scores = R_w @ S_items

# Cluster prior for each user (users × items)
cluster_prior = clust_item_prop.loc[d["Clusters"]].to_numpy()

# Recency weighting: more recent (smaller Recency) → higher weight
tau = 60.0
rec_w = np.exp(-d["Recency"].fillna(d["Recency"].median()) / tau).to_numpy().reshape(-1,1)

# Blend
alpha, beta = 0.7, 0.3  # CF vs cluster prior
scores_items = (alpha * cf_scores + beta * cluster_prior) * rec_w  # users × items

# Optional: avoid recommending the user's already-dominant category (encourage cross-sell)
user_item_share = d[item_cols].div(d[item_cols].sum(axis=1).replace(0,1), axis=0).fillna(0).to_numpy()
mask_dominant = user_item_share >= 0.50
scores_items = np.where(mask_dominant, -np.inf, scores_items)

# ---------- 3) Channel recommendation logic ----------
# User channel usage share
user_channel_share = d[channel_cols].div(d[channel_cols].sum(axis=1).replace(0,1), axis=0).fillna(0)

# For each user, base channel score = blend of (cluster channel preference) and (user’s own channel history)
# Also (optionally) boost "NumDealsPurchases" channel for promo-responsive users
cluster_channel_for_user = clust_channel_prop.loc[d["Clusters"]].reset_index(drop=True)
scores_channels = 0.6 * cluster_channel_for_user.values + 0.4 * user_channel_share.values  # users × channels

# Optional: promo boost for "NumDealsPurchases" if user historically responds to promos
if "NumDealsPurchases" in channel_cols and promo_cols:
    user_promo_intensity = d[promo_cols].sum(axis=1).fillna(0)  # higher => more responsive
    deals_idx = channel_cols.index("NumDealsPurchases")
    scores_channels[:, deals_idx] *= (1.0 + 0.15 * np.tanh(user_promo_intensity.to_numpy()))  # gentle boost

# Normalize channel scores per user for readability
scores_channels = scores_channels / (scores_channels.sum(axis=1, keepdims=True) + 1e-9)

# ---------- 4) Recommend for a user ----------
user_index = d.index.to_list()
items = item_cols
channels = channel_cols

def recommend_for_user(user_row_or_label, topn_items=3, topn_channels=2):
    # Resolve row index
    r = user_row_or_label if isinstance(user_row_or_label, int) else user_index.index(user_row_or_label)
    
    # Top-N items
    s_items = scores_items[r].copy()
    top_item_idx = np.argsort(-s_items)[:topn_items]
    rec_items = pd.DataFrame({
        "item": [items[i] for i in top_item_idx],
        "score": s_items[top_item_idx]
    })
    
    # Top channels
    s_ch = scores_channels[r].copy()
    top_ch_idx = np.argsort(-s_ch)[:topn_channels]
    rec_channels = pd.DataFrame({
        "channel": [channels[i] for i in top_ch_idx],
        "score": s_ch[top_ch_idx]
    })

    # Why (human-readable)
    reasons = {
        "cluster": int(d.iloc[r]["Clusters"]),
        "recency_days": int(d.iloc[r]["Recency"]),
        "cluster_top_items": top_items_for_cluster(int(d.iloc[r]["Clusters"]), topn=3).to_dict(),
        "cluster_top_channels": top_channels_for_cluster(int(d.iloc[r]["Clusters"]), topn=2).to_dict(),
        "user_top_spend": d[item_cols].iloc[r].sort_values(ascending=False).head(2).to_dict(),
        "user_channel_mix": d[channel_cols].iloc[r].to_dict()
    }
    return rec_items, rec_channels, reasons

# ---------- 5) Recommend per-cluster (for campaigns) ----------
def recommend_for_cluster(cluster_id, topn_items=5, topn_channels=3):
    mask = (d["Clusters"] == cluster_id).to_numpy()
    if mask.sum() == 0:
        return (pd.DataFrame(columns=["item","avg_score"]), pd.DataFrame(columns=["channel","avg_score"]))
    avg_item_scores = np.nanmean(scores_items[mask], axis=0)
    avg_channel_scores = np.nanmean(scores_channels[mask], axis=0)
    
    top_item_idx = np.argsort(-avg_item_scores)[:topn_items]
    top_ch_idx = np.argsort(-avg_channel_scores)[:topn_channels]
    
    items_df = pd.DataFrame({"item": [items[i] for i in top_item_idx],
                             "avg_score": avg_item_scores[top_item_idx]})
    channels_df = pd.DataFrame({"channel": [channels[i] for i in top_ch_idx],
                                "avg_score": avg_channel_scores[top_ch_idx]})
    return items_df, channels_df

# ---------- 6) Examples ----------
u0_items, u0_channels, why0 = recommend_for_user(user_row_or_label=0, topn_items=3, topn_channels=2)
print("User 0 — Recommended Items:\n", u0_items, "\n")
print("User 0 — Recommended Channels:\n", u0_channels, "\n")
print("Why:\n", why0, "\n")

for cid in sorted(d["Clusters"].unique()):
    ci_items, ci_channels = recommend_for_cluster(cid, topn_items=5, topn_channels=3)
    print(f"Cluster {cid} — Top Items:\n", ci_items, "\n")
    print(f"Cluster {cid} — Top Channels:\n", ci_channels, "\n")
```
